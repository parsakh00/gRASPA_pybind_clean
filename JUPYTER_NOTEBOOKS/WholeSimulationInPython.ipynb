```python
import gRASPA as g
```

We first create a simulation called "W", and run a simulation with 100 initialization cycles, for CO2 adsorption in MFI

```python
W = g.Initialize()
W.NumberOfInitializationCycles = 100
W.NumberOfProductionCycles = 0
g.RUN(W)
g.finalize(W)
```

We can check the final energy of this W simulation, system 0
* NOTE: The energies printed here are in internal unit of gRASPA (also in RASPA-2 and RASPA-3), 10 kJ/mol

```python
W_E = g.get_total_energy(W, 0)
```

Here We can try to **Reproduce** the simulation result for W by **Unwrapping** the simulation processes, running them in python

```python
V = g.Initialize()
V.NumberOfInitializationCycles = 100
V.NumberOfProductionCycles = 0
```

```python
# Declare MC Move classes
SINGLEPARTICLE = g.SingleMove()
INSERTION   = g.InsertionMove()
DELETION    = g.DeletionMove()
REINSERTION = g.ReinsertionMove()

# Initialize MC simulation for current simulation phase

box_index = 0
g.InitializeMC(V, box_index)

start = g.omp_get_wtime()


for i in range(0, V.NumberOfInitializationCycles):
  Steps = g.Determine_Number_Of_Steps(V, box_index, i)
  for s in range(0, Steps):
    #g.RunMoves(V, box_index, i)
    #'''
    g.Select_Box_Component_Molecule(V, box_index)
    # run single MC move #
    r = V.RandomNumber
    comp = V.SystemComponents[box_index].MCMoveVariables.component
    mol  = V.SystemComponents[box_index].MCMoveVariables.molecule

    # Select move #
    # for translation/rotation/deletion, if there is no molecule, skip #
    DeltaE = g.MoveEnergy()
    MoveType = g.MoveTypes.TRANSLATION
    if(r < V.SystemComponents[box_index].Moves[comp].TranslationProb):
      if(V.SystemComponents[box_index].NumberOfMolecules[comp] == 0): continue
      MoveType = g.MoveTypes.TRANSLATION; 
      V.SystemComponents[box_index].MCMoveVariables.MoveType = int(MoveType)
      DeltaE = g.SingleBodyMove(V, box_index)
    elif(r < V.SystemComponents[box_index].Moves[comp].RotationProb):
      if(V.SystemComponents[box_index].NumberOfMolecules[comp] == 0): continue
      MoveType = g.MoveTypes.ROTATION; 
      V.SystemComponents[box_index].MCMoveVariables.MoveType = int(MoveType)
      DeltaE = g.SingleBodyMove(V, box_index)
    elif(r < V.SystemComponents[box_index].Moves[comp].ReinsertionProb): 
      MoveType = g.MoveTypes.REINSERTION;
      V.SystemComponents[box_index].MCMoveVariables.MoveType = int(MoveType)
      if(V.SystemComponents[box_index].NumberOfMolecules[comp] == 0): continue
      DeltaE = REINSERTION.Run(V, box_index)
    elif(r < V.SystemComponents[box_index].Moves[comp].SwapProb):
      if(g.Get_Uniform_Random() < 0.5):
        #MoveType = g.MoveTypes.SINGLE_INSERTION
        MoveType = g.MoveTypes.INSERTION; 
        V.SystemComponents[box_index].MCMoveVariables.MoveType = int(MoveType)
        #DeltaE = g.SingleBodyMove(V, box_index) 
        DeltaE = INSERTION.Run(V, box_index);
      else:
        #MoveType = g.MoveTypes.SINGLE_DELETION
        MoveType = g.MoveTypes.DELETION; 
        V.SystemComponents[box_index].MCMoveVariables.MoveType = int(MoveType)
        if(V.SystemComponents[box_index].NumberOfMolecules[comp] == 0): continue
        DeltaE = DELETION.Run(V, box_index);
        #DeltaE = g.SingleBodyMove(V, box_index)
    g.MoveEnergy_Add(V.SystemComponents[box_index].deltaE, DeltaE)
    #'''
  # Gather statistics and averages per cycle #
  g.GatherStatisticsDuringSimulation(V, box_index, i)
g.MCEndOfPhaseSummary(V) # end of initialization phase summary #

end = g.omp_get_wtime()
print(f"simulation took {end - start} seconds\n")
g.finalize(V) # end of simulation #
```

We can check again for the total energy of simulation V, system 0

```python
V_E = g.get_total_energy(V, 0)
```

Calculate the difference between the total energies of W and V, we can see they are the same!
* Unwrapping the simulation does not break it and it still follows the same random number ordering!

```python
W_E.total() - V_E.total()
```
